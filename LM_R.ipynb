Linear Modeling R

```r
data = read.csv("regrex1.csv")
```

```r
head(data)
```

```r
Dependent_Variable = data$x
Independent_Variable = data$y
```

```r
plot(Dependent_Variable, Independent_Variable, main = "Cool Data")
```

```r
LM = lm(data)
```

```r
summary(LM)
```

```r
y_predict = predict(LM, data)
```

```r
plot(Dependent_Variable, Independent_Variable, main = "Cool Data")
lines(Dependent_Variable, y_predict)
```
